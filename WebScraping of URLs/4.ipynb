```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url="https://www.socialmediatoday.com/news/seo-copywriting-guide-10-tips-to-improve-your-content-infographic/625336/"
```

```python
response= requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
content = soup.find('div', {"class": "print-wrapper"})
```

```python
article = ''
for i in content.findAll('div'):
    article = article + ' ' +  i.text
print(article)
```

```python
with open('4.txt', 'w') as file:
    file.write(article)
```
