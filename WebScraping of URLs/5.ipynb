```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url="https://mobileki.site/mobile-phones-under-15000"
```

```python
response= requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
content = soup.find('div', {"class": "col-12 col-md-9"})
```

```python
article = ''
for i in content.findAll('div'):
    article = article + ' ' +  i.text
print(article)
```

```python
with open('5.txt', 'w') as file:
    file.write(article)
```
