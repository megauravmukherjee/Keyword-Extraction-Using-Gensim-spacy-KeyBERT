```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url="https://www.gonnawantseconds.com/italian-wedding-cookies/"
```

```python
response= requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
content = soup.find('div', {"class": "site-container"})
```

```python
article = ''
for i in content.findAll('div'):
    article = article + ' ' +  i.text
print(article)
```

```python
#with open('6.txt', 'w') as file:
  #  file.write(article)
    
with open("6.txt", 'wb') as file:
        file.write(article.encode('utf-8', 'ignore'))
```
