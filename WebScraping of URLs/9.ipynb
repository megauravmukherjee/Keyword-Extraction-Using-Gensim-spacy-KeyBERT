```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url="https://jupiter.money/international-money-transfer/best-apps-to-send-money-abroad/"
```

```python
response= requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
content = soup.find('div', {"class": "max-w-screen-md mx-auto"})
```

```python
article = ''
for i in content.findAll('div'):
    article = article + ' ' +  i.text
print(article)
```

```python
with open('9.txt', 'w') as file:
    file.write(article)
```
