```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url="https://getintopc.com/softwares/file-sharing/vandyke-securecrt-and-securefx-2021-free-download/"
```

```python
response= requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
content = soup.find('div', {"class": "blocks clear-block"})
```

```python
article = ''
for i in content.findAll('div'):
    article = article + ' ' +  i.text
print(article)
```

```python
with open('7.txt', 'w') as file:
    file.write(article)
```
